# Loading data â­•

```python
data = pd.read_csv('../input/marvel-vs-dc/db.csv', encoding = 'latin')
```

## Removing columns that do not add any importance to analyse the dataset

```python
data = data.drop(['Unnamed: 0', 'Original Title'], axis = 1)
```

```python
data
```

```python
data.info()
```

> All columns except 'Company' type should be converted to either float or int type

```python
convert_dtype = {
    'Minutes' : int,
    'Budget' : int
}

data = data.astype(convert_dtype)
data.info()
```

> Converted

```python
data[data['Company'] == 'Marvel'].mean()
```

```python
data[data['Company'] == 'DC'].mean()
```

# Graphs showing correlation ðŸ“ˆ

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Company', y = 'Rate', data = data)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Company', y = 'Metascore', data = data)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Company', y = 'Minutes', data = data)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Release', y = 'Rate', hue = 'Company', data = data)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Release', y = 'Metascore', hue = 'Company', data = data)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Release', y = 'Minutes', hue = 'Company', data = data)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Company', y = 'Budget', data = data)
plt.show()
```

```python
plt.figure(figsize = (12, 10))
sns.barplot(x = 'Company', y = 'Gross Worldwide', data = data)
plt.show()
```
